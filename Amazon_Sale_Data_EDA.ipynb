#### imputation technique

Replace null values by mean,median or mode technique

Mean--> Average. When data is numerical then mean imputation technique is used.
Median--> Middle most value.When data is numerical then mean imputation technique is used.
Mode--> Mostly occuring data points.When data is categorical then mode imputation technique is used.

```python
df['currency'].unique()
```

```python
df['ship-country'].unique()
```

```python
df['fulfilled-by'].unique()
```

#### Mean imputation technique

```python
df['Amount'].fillna(df['Amount'].mean(),inplace=True)
```

```python
# check Amount column having zero null value
```

```python
df['Amount'].isnull().sum()
```

```python
df.columns
```

#### Mode imputation technique

```python
df['currency'].fillna(df['currency'].mode()[0],inplace=True)
```

```python
df['currency'].isnull().sum()
```

```python
df['ship-city'].fillna(df['ship-city'].mode()[0],inplace=True)
df['ship-state'].fillna(df['ship-city'].mode()[0],inplace=True)
df['ship-postal-code'].fillna(df['ship-postal-code'].mode()[0],inplace=True)
df['ship-country'].fillna(df['ship-country'].mode()[0],inplace=True)
df['fulfilled-by'].fillna(df['fulfilled-by'].mode()[0],inplace=True)
```

```python
top_10_state=df['ship-state'].value_counts().head(10)
```

```python
top_10_state
```

```python
a=top_10_state.plot(kind='bar',color='g',edgecolor='r')
for i in a.containers:
    a.bar_label(i)
plt.title('Top 10 state based on product purches')
plt.xlabel('State Name')
plt.ylabel('Count of Produts')
```

```python
top_10_city=df['ship-city'].value_counts().head(10)
top_10_city
```

```python
a=top_10_city.plot(kind='bar',color='g',edgecolor='r')
for i in a.containers:
    a.bar_label(i)
plt.title('Top 10 city based on product purches')
plt.xlabel('City Name')
plt.ylabel('Count of Produts')
```
